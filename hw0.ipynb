```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
from math import pi
from math import sqrt
from math import sin
from scipy import special
```

```python
a = 1.5 + 15j

def f(x):
    return np.exp(a * 1j * x)

def lnf(x):
    #print(1j * a * x)
    return 1j * a * x

def taylor(x):
    y = []
    for i in range(num):
        estVal = 0.0
        for j in range(N):
            estVal += (1j * a * x[i]) ** j /factorial(j)
            #print(np.log(estVal))
        y.append(estVal)
        #print(np.log(estVal), x[i])
    return y
```

```python
num = 1000
N = 140
inputs = np.linspace(-10, 10, num)
fval = f(inputs)
analyticVal = lnf(inputs)
intrinsicVal = np.log(fval)
expVal = np.log(taylor(inputs))
```

```python
plt.figure(figsize = (10, 8))
plt.xticks(np.arange(-10, 10, 1))
plt.plot(inputs, analyticVal.real, label = 'Real part')
plt.plot(inputs, analyticVal.imag, label = 'Imaginary part')
plt.title('lnf(x) analytically')
plt.legend()
plt.show()
```

```python
def ellpint(k, iter_power):
    integral_value = 0.0
    for i in range(iter_power):
        integral_value += pi/2 * (factorial(2 * i) / (4 ** i * (factorial(i) ** 2))) ** 2 * (k ** (2*i))
    return integral_value
```

```python
print(ellpint(1/sqrt(2), 1000))
```

```python
def ellipf1(x):
    return (1-(sin(x))**2 * 0.5) ** (-1/2)
def ellipf2(x):
    return ((1-x ** 2) * (1-0.5 * x ** 2)) ** (-1/2)

def midpoint_1(steps):
    integral_value = 0.0
    startpoint = 0.0
    endpoint = pi/2
    interval = endpoint / steps
    midpoint = startpoint + 0.5 * interval
    for i in range(steps):
        integral_value += ellipf1(midpoint) * interval
        midpoint += interval
    return integral_value

def midpoint_2(steps):
    integral_value = 0.0
    startpoint = 0.0
    endpoint = 1
    interval = endpoint / steps
    midpoint = startpoint + 0.5 * interval
    for i in range(steps):
        integral_value += ellipf2(midpoint) * interval
        midpoint += interval
    return integral_value    
```

```python
steps = 20000
list_m1 = []
list_m2 = []
list_steps = []
for i in range(1, steps):
    list_m1.append(midpoint_1(i))
    list_steps.append(i)
    list_m2.append(midpoint_2(i))
plt.figure(figsize = (10, 10))
plt.ylim(1.84, 1.855)
plt.yticks(np.arange(1.84, 1.855, 0.001)) 
plt.plot(list_steps, list_m1, label = 'First form')
plt.plot(list_steps, list_m2, label = 'Second form')
plt.legend()
plt.show()
```
