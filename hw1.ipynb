```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
from math import pi
from math import sqrt
from math import sin
from scipy import special
from numpy import exp
```

```python
steps = 1000

def f(episilon, omega):
    delta = 1e-10
    return sqrt(4 - episilon ** 2)/(omega - episilon + delta * 1j)/2/pi

def g(episilon):
    return 1/sqrt((4 - episilon ** 2))/2/pi

def G(omega):
    integral_value = 0.0
    startpoint = -2.0
    endpoint = 2.0
    interval = endpoint / steps
    midpoint = startpoint + 0.5 * interval
       
    for i in range(steps):
        if abs(omega) > 2:
            integral_value += f(midpoint, omega) * interval
        else:
            integral_value += (g(omega) + 1j * (-1)/2 * sqrt(4 - omega ** 2)) * interval / 2
        midpoint += interval
    return integral_value
```

```python
inputs = np.linspace(-5, 5, steps)
ylist = []
for i in range(len(inputs)):
    ylist.append(G(inputs[i]))
```

```python
fval = np.asarray(ylist)
```

```python
plt.figure(figsize = (10, 8))
#plt.xticks(np.arange(-10, 10, 1))
plt.ylabel(r'$G(\omega)$')
plt.xlabel(r'$\omega$')
plt.plot(inputs, fval.real, label = 'Real part')
plt.plot(inputs, fval.imag, label = 'Imaginary part')
plt.legend()
plt.show()
```
